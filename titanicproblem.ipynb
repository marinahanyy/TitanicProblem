```python
cols = ['Name', 'Ticket', 'Cabin']
train_data = train_data.drop(cols, axis=1)
train_data.info()
train_data.head()
```

Here i dropped the features Name,ticket and cabin from train_data file as they will not be useful in my submission. Steps: Dtermining the names of columns needed to be dropped then drop them. Conclusion: Dropping the needed columns from the data file as shown above.

```python
train_data['Age'] = train_data['Age'].interpolate()
train_data['Fare'] = train_data['Fare'].interpolate()
train_data['Fare'].fillna(train_data['Fare'],inplace= True)
train_data['Age'].fillna(train_data['Age'],inplace= True)
train_data['Embarked'].fillna("Unavailable",inplace= True)
print(train_data)
```

Here i make interpolation and fillna to Age and Fare to fill the missing values then i used fillna in Embarked to write in the NaN items unavailable. Steps: Interpolate both age and fare by interpolation method then fillna inside each one of them that fill the null values with the value of interpolation of each of them.

```python
train_data['Sex'] = preprocessing.LabelEncoder().fit_transform(train_data['Sex'])
train_data['Embarked'] = preprocessing.LabelEncoder().fit_transform(train_data['Embarked'])
train_data[['Sex','Embarked']].info()
train_data.head()
```

Here i converted all letters to numbers by using labelencoder.fittransform to sex and Embarked.
Steps:Using the method labelencoder to sex and Embarked. Conclusion: the letters in each are transformed to numbers as shown above.

```python
test_data['Age'] = test_data['Age'].interpolate()
test_data['Fare'] = test_data['Fare'].interpolate()
test_data['Fare'].fillna(test_data['Fare'],inplace= True)
test_data['Age'].fillna(test_data['Age'],inplace= True)
test_data['Embarked'].fillna("Unavailable",inplace= True)
print(test_data)
```

```python
y = train_data["Survived"]
X = train_data.drop("Survived",axis=1)
x_train,x_validation,y_train,y_validation=train_test_split(X,y,test_size=0.25,random_state=0)
```

Here i dropped column Survived from X and entered X as input and y is the output then i did the validation using train test split. Steps: Putting test_size= 0.25 and random_state=0.

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
```

Then here i used the model RandomForest giving it the parameters needed

```python
X_test = test_data
print(X_test)
model.fit(x_train, y_train)
Score=model.score(x_validation,y_validation)
print(Score)
predictions = model.predict(X_test)
```

Here i printed the score and the accuarcy of the prediction of the model giving the score validation in parameters.Steps:putting test file in X_test and give the model.train the x_tain and y_train Conclusion: then calculating the score and printing the accuracy of the predictions.
